```python
#Histogram plot showing the distribution of Calories of items in McDonalds menu
plt.figure(figsize=(10,6))
sns.histplot(df['Calories'], kde=True, bins=30)
plt.title('Distribution of calories of items in McDonalds menu')
plt.xlabel('calories')
plt.ylabel('frequency')
plt.show()
```

```python
#Observations
#More than 30 items in the McDonalds menu have the Calories inthe range 300
#More than 25 items in menu have Calories less than 250
#More than 25 items in the menu have Calories more than 500
#Most of the items have calories range between 150 to 550
```

```python
#Nutritional Content Analysis
# Distribution of Total Fat across the menus
plt.figure(figsize = (10,6))
sns.histplot(df['Total Fat'], kde=True, bins=20)
plt.title('Distribution of Total fat in McDonalds menu item')
plt.xlabel('Total fat(g)')
plt.ylabel('Frequency')
plt.show()
```

```python
#Observations
#More than 80 items in the menu have Total fat less than 5g
#More than 40 item in the menu have Total fat between 10g to 20g
```

```python
unique = df['Category'].unique()
unique
```

```python
#Average calories per category
avg_calorie = df.groupby('Category')['Calories'].mean()
avg_calorie
```

```python
#Observations
#In the category part 'Smoothies & Shakes' has highest calories and Bevarages have the less amount of calories
```

```python
avg_calorie = avg_calorie.reset_index
```

```python

# Group by 'Category' and calculate the mean of 'Calories'
avg_calorie = df.groupby('Category')['Calories'].mean().reset_index()

# Plotting the barplot
plt.figure(figsize=(6, 4))
sns.barplot(x='Calories', y='Category', data=avg_calorie)
plt.title('Average Calories by Category')
plt.xlabel('Category')
plt.ylabel('Average Calories')
plt.show()
```

```python
#Displaying the Boxplot of different Categories 
plt.figure(figsize=(14, 8))
sns.boxplot(x='Calories', y='Category', data=df)
plt.xticks(rotation=45)
plt.title('Calories by Category')
plt.xlabel('Average calories')
plt.ylabel('Category')
plt.show()
```

```python
# Pairplot to compare nutritional components
sns.pairplot(df[['Calories', 'Total Fat', 'Saturated Fat', 'Carbohydrates', 'Protein']], diag_kind='kde')
plt.suptitle('Pairplot of Nutritional Components', y=1.02)
plt.show()
```

```python
#Observations
#Pairplot shows relationship between two variables
# it also displays the Distribution of the curve
#By seeing the pairplots it is known that there is a good relationship between
#i) Total Fat and Calories, ii)Saturation fat and Total Fat, iii)Calories and Carbohydrates
# The Kde line graph shows that all the the line plot is Right skeiwed
#That means Higher values and outliers are present
```

```python
print('Top 10 highest calories')
print(df.nlargest(10, 'Calories')[['Item', 'Calories', 'Category']])

print('Top 10 smallest calories')
print(df.nsmallest(10, 'Calories')[['Item', 'Calories', 'Category']])
```

```python
# Average nutritional content by category
category_means = df.groupby('Category').mean(numeric_only =True)
print(category_means[['Calories', 'Total Fat', 'Protein', 'Carbohydrates']])
```
